# Gram staining manipulation

```python
def get_genus(df: pd.DataFrame):
    df['Genus'] = df['species'].apply(lambda x: x.split()[0])
    return df
```

```python
positive = pd.read_csv("../data/external/positive.csv", skiprows=1)
positive = positive.iloc[:, -2:]
positive = get_genus(positive)
print(positive.shape)
positive.head()
```

```python
negative = pd.read_csv("../data/external/negative.csv", skiprows=1)
negative = negative.iloc[:, -2:]
negative = get_genus(negative)
print(negative.shape)
negative.head()
```

```python
variable = pd.read_csv("../data/external/variable.csv", skiprows=1)
variable = variable.iloc[:, -2:]
variable = get_genus(variable)
print(variable.shape)
variable.head()
```

```python
# Get unique genera from each DataFrame
unique_genus_df2 = set(positive['Genus'])
unique_genus_df3 = set(negative['Genus'])
unique_genus_df4 = set(variable['Genus'])

# Find common genera between pairs of DataFrames
common_genus_df2_df3 = unique_genus_df2.intersection(unique_genus_df3)
common_genus_df2_df4 = unique_genus_df2.intersection(unique_genus_df4)
common_genus_df3_df4 = unique_genus_df3.intersection(unique_genus_df4)

# Check and display where common genera exist
if common_genus_df2_df3:
    print(f"Common genera between df2 and df3: {common_genus_df2_df3}")

if common_genus_df2_df4:
    print(f"Common genera between df2 and df4: {common_genus_df2_df4}")

if common_genus_df3_df4:
    print(f"Common genera between df3 and df4: {common_genus_df3_df4}")
```

# Fixing duplicated species

```python
stain = {'Actinotalea ferrariae':'positive', 'Tistlia consotensis': 'positive', 'Belliella pelovolcani': 'negative', 
              'Ureibacillus massiliensis':'positive', 'Vallitalea guaymasensis':'positive', 'Clostridioides difficile':'positive'}

# Iterate through the dictionary and update the DataFrame
for species, gram_stain in stain.items():
    # Check if the species exists in the DataFrame
    if species in all_gram['species'].values:
        # Update the Gram stain value in the DataFrame
        all_gram.loc[all_gram['species'] == species, 'Gram stain'] = gram_stain
```

```python
# Before fixing
# List of species you want to filter by
species_list = ['Actinotalea ferrariae', 'Tistlia consotensis', 'Belliella pelovolcani', 'Ureibacillus massiliensis', 'Vallitalea guaymasensis', 'Clostridioides difficile']

# Filter the DataFrame to obtain rows where 'species' is in the list
all_gram[all_gram['species'].isin(species_list)].head()
```
